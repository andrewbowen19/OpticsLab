```python
# Reading in txt file with counts
#path= "C:\\Lab\Data\2020\Jan_13\\"
filename = 'Jan_15_2020_1-49PM_t_res_g2.txt'
oldpath = '/Users/andrewbowen/physics/OpticsLab/data/'
newpath = oldpath + filename


# newpath = os.path.join(os.path.join(os.path.join('C:\\Users', 'fmb1744'), 'Desktop'), filename)
# junk = pd.read_table(newpath, sep = '\t', skiprows =1,nrows = 8) #reading in first rows
dat = pd.read_table(newpath, sep = '\t', skiprows = 8)#might need names param, play with read-in

# print(dat)
newColNames = ['A', 'B', 'A\'', 'B\'', 'AB', 'ABB\'',
               'AB\'', 'BB\'', 'TAB', 'TAB\'', 'TBB\'', 'junk']
dat.columns = newColNames
```

```python
# Functions for difference calculation
def getDiff(A, B, Aprime, Bprime):
    '''Count Difference function'''
    diff = A - B + Aprime - Bprime
    print('difference: ', diff)
    return diff

def percentError(diff, B, Bprime):
    '''Percent error function'''
    err = diff / (B + Bprime)
    print('% Error: ', err * 100)
    return err
```

```python
A = dat['A'].values
B = dat['B'].values
Aprime = dat['A\''].values
Bprime = dat['B\''].values

xx = getDiff(A, B, Aprime, Bprime) # raw diff call
yy = percentError(abs(xx), B, Bprime) # percent error call (using abs for positive value)
zz = yy * 100 # converting to true percent
avgPerr = np.mean(zz)
print('Average % error: ', avgPerr)
```

```python
def g2_naive(a, ab, ab_prime, abb_prime): # only use when no beam splitter in the A-beam (Lower A' values)
    return a * abb_prime / (ab * ab_prime)
```

```python
#x and y are two-element lists, with first entry the value and second entry the error
def times_err(x, y):
    """Multiplication error propagation"""
    return [x[0] * y[0], np.sqrt((x[0] * y[1])**2 + (y[0] * x[1])**2)]

def plus_err(x, y):
    """Addition Error propagation"""
    return [x[0] + y[0], np.sqrt(x[1]**2 + y[1]**2)]
```

```python
# Getting g2_naive values - first setting up combination arrays
AB = dat['AB'].values
ABprime = dat['AB\''].values
AB_Bprime = dat['ABB\''].values

g2 = g2_naive(A, AB, ABprime, AB_Bprime)

print(g2)

```
