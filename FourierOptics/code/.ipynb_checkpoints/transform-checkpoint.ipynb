```python
# Notebook for computing the fourier transform of an image
import imageio as im
import matplotlib.pyplot as plt
import numpy as np

path = '/Users/andrewbowen/OpticsLab/FourierOptics/pics/'
```

```python
#define functions
def amplitude(fourier):
    '''Function that returns the amplitude of a fourier series'''
    amplitude=np.abs(fourier)
    return amplitude 

def phase(fourier):
    '''Function to return the phase of a fourier transform'''
    phase = np.ones(fourier.shape,dtype=complex)
    for i in range(fourier.shape[0]):
        for k in range(fourier.shape[1]):
            phase[i,k]=(1j)*np.angle(fourier[i,k])
    return phase 

# ##############################

def makeImages(subdir, filename1, filename2):
    '''Function to read in images and show them'''
    # reading in image files (located in our pics directory)
    image = im.imread(path + subdir + filename1)
    image2 = im.imread(path + subdir + filename2)
    image = np.dot(image[:,:,:3],[0.299,0.587,0.115])#image[:,:,0]
    image2 = np.dot(image2[:,:,:3],[0.299,0.587,0.114])
    print(image.shape)
    print(image2.shape)

    # 'Plotting' images to screen
    plt.imshow(image,cmap = 'gray')
    plt.title(filename1.replace('.jpg', '')'Image 1 - Read in')
    plt.show()
    plt.imshow(image2,cmap = 'gray')
    plt.title(filename2.replace('.jpg', '') + 'Image 2 - Read in')
    plt.show()

    #perform transform
    fourier1 = np.fft.rfft2(image)
    fourier2 = np.fft.rfft2(image2)
    print(fourier1.shape)
    print(fourier2.shape)

    print('Fourier reverse calculated...')

    #calculate phases and amplitudes
    phases1 = phase(fourier1)
    amplitudes1 = amplitude(fourier1)

    phases2 = phase(fourier2)
    amplitudes2 = amplitude(fourier2)

    print('Amplitudes and phases done!')

    # merge phases and amplitudes
    mix12 = np.multiply(amplitudes1,np.exp(phases2))
    mix21 = np.multiply(amplitudes2,np.exp(phases1))

    #conduct reverse transform
    mimage12 = np.fft.irfft2(mix12)
    mimage21 = np.fft.irfft2(mix21)

    print('Creating images...')

    #show images
    plt.imshow(mimage12,cmap = 'gray')
    plt.title('Image 1 - FT')
    plt.show()
    plt.imshow(mimage21,cmap = 'gray')
    plt.title('Image 2 - FT')
    plt.show()

    return image, image2

image, image2 = makeImages('stripes/','ElderWang.JPG', 'inverseElderWang.pdf')
```
