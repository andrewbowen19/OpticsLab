```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit
```

```python
path = '/Users/andrewbowen/OpticsLab/Interferometry/indexOfRefraction/dataIndex/'
dataAir = pd.read_csv(path + 'Gas-Index-Refraction.csv', sep = ',')

def cumTotal(df):
    new_counts = []
    c = 0
    for index,row in df.iterrows():
        net = abs(row['Counts (N)'])# - 100000)
        c += net
        new_counts.append(c)
    df['new counts'] = new_counts
        
    return df
    


dataAir

cumTotal(dataAir)
```

```python
# Defining the functions we'll need; maybe write this up as a class
def linModel(x,m,b):
    return m*x + b

def getIndex(N):
    '''Function that returns the index of refraction from the # of counts we get on the microcontroller
        From Teachspin Manual pg 65 (bottom)'''
    wavelength = 632.990 * (10**-9)
    L = 10.000
    n = ((N * wavelength)/(2 * L)) + 1
    return n

def linearFitPlot(x,y, gasType):
    # Fitting the curves
    fitCoeffs, fitCovariances = curve_fit(linModel, x, y)

    # Pick out the best-fit for the 'm' value and best-fit for the 'b' value
    bestfit_m = fitCoeffs[0] # This slope is related to coeff of thermal expansion
    bestfit_b = fitCoeffs[1]

    # Printing best-fit values to screen, bestfit_m is slope of linear fit, b is y-intercept
    print(f'best-fit value for \'m\': {bestfit_m}')
    print(f'best-fit value for \'b\': {bestfit_b}')
    
    bestfit_x = np.arange(np.max(x))
#     Define the y-values for the best-fit line, using the linModel function defined above
    bestfit_y = linModel(bestfit_x, bestfit_m, bestfit_b)
    
    # Plot your data (again) as red circles with error bars
    f,ax = plt.subplots(figsize=(8,5))
    ax.scatter(x,y, c = 'r')
#     ax.scatter(indices[1:],voltage[1:], c = 'r')
    ax.set_xlabel(f"Index of refraction: {gasType}", fontsize = 16)
    ax.set_ylabel("Voltage (V)", fontsize = 16)
    ax.set_title(gasType)


    # # Overplot the best-fit line in blue (default color)
#     ax.plot(bestfit_x, bestfit_V)
```

```python
# Need to incorporate calculations from notes on canvas
L = 10e-7# length of chamber - 10cm form TeachSpin
countsAir = dataAir['Counts (N)']
voltageAir = dataAir['Voltage (V)']

indicesAir = getIndex(countsAir)
print('Indices of refraction for Air:')
print(indicesAir)
```

```python
linearFitPlot(indicesAir, voltageAir, 'Air')
```

```python
time = dataVoltTime['time (s)']
volts = dataVoltTime['Voltage (V)']
```

```python
# Plotting voltage time series - checking to make sure we can reach equilibrium within 10 min easily
f,ax = plt.subplots(figsize = (8,5))
ax.scatter(time, volts)
ax.set_xlabel('time (s)')
ax.set_ylabel('Voltage (V)')
```

```python
def logModel(N0, t, d):
    return -N0 * np.log(d * t)

# N0 - initial value
# d = lambda where lambda is the decay param
```

```python
# Fitting the curves
fitCoeffs, fitCovariances = curve_fit(logModel, time, volts)
print(fitCoeffs)
# Pick out the best-fit for the 'm' value and best-fit for the 'b' value
bestfit_N0 = fitCoeffs[0] 
bestfit_d = fitCoeffs[1]



# Printing best-fit values to screen, bestfit_m is slope of linear fit, b is y-intercept
print(f'best-fit value for \'N0\': {bestfit_N0}')
print(f'best-fit value for \'d\': {bestfit_d}')
print(f'best-fit value for \'b\': {bestfit_b}')
```

```python
bestfit_t = np.arange(0,230)
bestfit_V = logModel(bestfit_N0, bestfit_t, bestfit_d)
# print(bestfit_V)
```

```python
f, ax = plt.subplots()
ax.scatter(time, volts)
ax.set_xlabel('time (s)')
ax.set_ylabel('Voltage (V)')

ax.plot(bestfit_t, bestfit_V)
```
